### 1.	Validando a consistência dos dados

```python
quarto = df.comodos.value_counts().sort_index()
comodo = []
quantidade_comodo = []

for i,v in quarto.items():
    if i not in comodo:
        comodo.append(i)
    if v not in quantidade_comodo:
        quantidade_comodo.append(v)

print(comodo)
print(quantidade_comodo)

plt.bar(comodo,quantidade_comodo)
plt.grid()
plt.show()

```

```python
banheiro = df.banheiros.value_counts()

banheiros = []
quant_banheiro = []

for i,v in banheiro.items():
    if i not in banheiros:
        banheiros.append(i)
    if v not in quant_banheiro:
        quant_banheiro.append(v)

print(banheiros)
print(quant_banheiro)

plt.bar(banheiros,quant_banheiro)
plt.grid()
plt.show()
```

```python

valores = []
cidades = []
city = df.cidade.value_counts()

for i,v in city.items():
    if i not in cidades:
        cidades.append(i)
    if v not in valores:
        valores.append(v)
print(valores)
print(cidades)

plt.bar(cidades,valores)
plt.grid()
plt.show()
    
```

```python
pet = df.pets.value_counts()
aceita_ou_nao_pet = []
quant_pet = []

for i,v in pet.items():
    if i not in aceita_ou_nao_pet:
        aceita_ou_nao_pet.append(i)
    if v not in quant_pet:
        quant_pet.append(v)

print(aceita_ou_nao_pet)
print(quant_pet)

plt.bar(aceita_ou_nao_pet,quant_pet)
plt.grid()
plt.show()

```

```python
mobilia = df.mobilia.value_counts()
mobilia_ou_nao = []
quant_mobilia = []

for i,v in mobilia.items():
    if i not in mobilia_ou_nao:
        mobilia_ou_nao.append(i)
    if v not in quant_mobilia:
        quant_mobilia.append(v)

print(mobilia_ou_nao)
print(quant_mobilia)

plt.bar(mobilia_ou_nao,quant_mobilia)
plt.grid()
plt.show()
```

```python
vaga = df.vagas.value_counts()
vagas = []
quant_vagas = []

for i,v in vaga.items():
    if i not in vagas:
        vagas.append(i)
    if v not in quant_vagas:
        quant_vagas.append(v)

print(vagas)
print(quant_vagas)

plt.bar(vagas,quant_vagas)
plt.grid()
plt.show()
```

```python
andar = df.andar.value_counts().sort_index()
andares = []
quant_andares = []

for i,v in andar.items():
    if i not in andares:
        andares.append(i)
        quant_andares.append(v)

print(andares)
print(quant_andares)

plt.bar(andares,quant_andares)
plt.grid()
plt.show()
```

```python
df_area = df.area

plt.boxplot(df_area)
plt.grid()
plt.show()
```

```python
df_condominio = df.condominio

plt.boxplot(df_condominio)
plt.grid()
plt.show()
```

```python
df_IPTU = df.IPTU

plt.boxplot(df_IPTU)
plt.grid()
plt.show()
```

```python
de_seguro = df.seguro

plt.boxplot(de_seguro)
plt.grid()
plt.show()
```

```python
df_aluguel = df.aluguel

plt.boxplot(df_aluguel)
plt.grid()
plt.show()

df.head()
```

```python
# SUBPLOT DE TODOS OS GRAFICOS

plt.figure(figsize = (23,15))
plt.subplot(4,3,1)

plt.bar(cidades,valores)
plt.title("Barras das Cidades")
plt.grid()

plt.subplot(4,3,2)
plt.boxplot(df_area)
plt.title("boxplot da área")
plt.grid()

plt.subplot(4,3,3)
plt.bar(comodo,quantidade_comodo)
plt.title("Barras dos comodos")
plt.grid()

plt.subplot(4,3,4)
plt.bar(banheiros,quant_banheiro)
plt.title("Barras dos banheiros")
plt.grid()

plt.subplot(4,3,5)
plt.bar(vagas,quant_vagas)
plt.title("Barras das vagas")
plt.grid()

plt.subplot(4,3,6)
plt.bar(andares,quant_andares)
plt.title("Barras dos andares")
plt.grid()

plt.subplot(4,3,7)
plt.bar(aceita_ou_nao_pet,quant_pet)
plt.title("Barras dos pets")
plt.grid()

plt.subplot(4,3,8)
plt.bar(mobilia_ou_nao,quant_mobilia)
plt.title("Barras das mobilias")
plt.grid()

plt.subplot(4,3,9)
plt.boxplot(df_condominio)
plt.title("Boxplot do condominio")
plt.grid()

plt.subplot(4,3,10)
plt.boxplot(df_IPTU)
plt.title("Boxplot do IPTU")
plt.grid()

plt.subplot(4,3,11)
plt.boxplot(de_seguro)
plt.title("Boxplot do seguro")
plt.grid()

plt.subplot(4,3,12)
plt.boxplot(df_aluguel)
plt.title("Boxplot do aluguel")
plt.grid()

```

### 2.	Analisar relação entre as variáveis explicativas qualitativas com preço do aluguel 

```python
al_med = df.groupby('cidade')['aluguel'].mean()

seg_med = df.groupby('cidade')['seguro'].mean()

iptu_med = df.groupby('cidade')['IPTU'].mean()

cond_med = df.groupby('cidade')['condominio'].mean()

area_med = df.groupby('cidade')['area'].mean()

media_por_cidade = pd.DataFrame({
    'Aluguel Médio': al_med,
    'Seguro Médio': seg_med,
    'IPTU Médio': iptu_med,
    'Condomínio Médio': cond_med,
    'Area Média': area_med
})
```

```python
media_por_cidade['Aluguel Médio por m²'] = media_por_cidade['Aluguel Médio'] / media_por_cidade['Area Média']
media_por_cidade['Seguro Médio por m²'] = media_por_cidade['Seguro Médio'] / media_por_cidade['Area Média']
media_por_cidade['IPTU Médio por m²'] = media_por_cidade['IPTU Médio'] / media_por_cidade['Area Média']
media_por_cidade['Condomínio Médio por m²'] = media_por_cidade['Condomínio Médio'] / media_por_cidade['Area Média']

media_por_cidade['Total Médio por m²'] = (
    media_por_cidade['Aluguel Médio por m²'] +
    media_por_cidade['Seguro Médio por m²'] +
    media_por_cidade['IPTU Médio por m²'] +
    media_por_cidade['Condomínio Médio por m²']
)
```

```python
max_total_medio_por_m2 = media_por_cidade['Total Médio por m²'].max()
cidade_max_total_medio_por_m2 = media_por_cidade[media_por_cidade['Total Médio por m²'] == max_total_medio_por_m2].index[0]

print(f'O lugar mais caro para se viver é {cidade_max_total_medio_por_m2} com um valor de médio de R$ {max_total_medio_por_m2:.2f} por m²')
```

```python
al_med = df.groupby(['cidade','mobiliado'])['aluguel'].mean()

seg_med = df.groupby(['cidade','mobiliado'])['seguro'].mean()

iptu_med = df.groupby(['cidade','mobiliado'])['IPTU'].mean()

cond_med = df.groupby(['cidade','mobiliado'])['condominio'].mean()

area_med = df.groupby(['cidade','mobiliado'])['area'].mean()

media_por_cidade = pd.DataFrame({
    'Aluguel Médio': al_med,
    'Seguro Médio': seg_med,
    'IPTU Médio': iptu_med,
    'Condomínio Médio': cond_med,
    'Area Média': area_med
})

display(media_por_cidade)
```

```python
media_por_cidade['Aluguel Médio por m²'] = media_por_cidade['Aluguel Médio'] / media_por_cidade['Area Média']
media_por_cidade['Seguro Médio por m²'] = media_por_cidade['Seguro Médio'] / media_por_cidade['Area Média']
media_por_cidade['IPTU Médio por m²'] = media_por_cidade['IPTU Médio'] / media_por_cidade['Area Média']
media_por_cidade['Condomínio Médio por m²'] = media_por_cidade['Condomínio Médio'] / media_por_cidade['Area Média']

media_por_cidade['Total Médio por m²'] = (
    media_por_cidade['Aluguel Médio por m²'] +
    media_por_cidade['Seguro Médio por m²'] +
    media_por_cidade['IPTU Médio por m²'] +
    media_por_cidade['Condomínio Médio por m²']
)

display(media_por_cidade)
```

```python

media_nao_mobiliado = media_por_cidade.loc[(slice(None), 'nao mobiliado'), 'Total Médio por m²'].mean() 
print(f'Total Médio por m² para imóveis não mobiliados é {media_nao_mobiliado:.2f}')

media_mobiliado = media_por_cidade.loc[(slice(None), 'mobiliado'), 'Total Médio por m²'].mean()
print(f'Total Médio por m² para imóveis mobiliados é {media_mobiliado:.2f}')

print(f'A diferença percentual entre os dois valores é de {((media_mobiliado - media_nao_mobiliado) / media_nao_mobiliado) * 100:.2f}%')
```

```python
al_med = df.groupby(['cidade','pets'])['aluguel'].mean()

seg_med = df.groupby(['cidade','pets'])['seguro'].mean()

iptu_med = df.groupby(['cidade','pets'])['IPTU'].mean()

cond_med = df.groupby(['cidade','pets'])['condominio'].mean()

area_med = df.groupby(['cidade','pets'])['area'].mean()

media_por_cidade = pd.DataFrame({
    'Aluguel Médio': al_med,
    'Seguro Médio': seg_med,
    'IPTU Médio': iptu_med,
    'Condomínio Médio': cond_med,
    'Area Média': area_med
})

display(media_por_cidade)
```

```python
media_por_cidade['Aluguel Médio por m²'] = media_por_cidade['Aluguel Médio'] / media_por_cidade['Area Média']
media_por_cidade['Seguro Médio por m²'] = media_por_cidade['Seguro Médio'] / media_por_cidade['Area Média']
media_por_cidade['IPTU Médio por m²'] = media_por_cidade['IPTU Médio'] / media_por_cidade['Area Média']
media_por_cidade['Condomínio Médio por m²'] = media_por_cidade['Condomínio Médio'] / media_por_cidade['Area Média']

media_por_cidade['Total Médio por m²'] = (
    media_por_cidade['Aluguel Médio por m²'] +
    media_por_cidade['Seguro Médio por m²'] +
    media_por_cidade['IPTU Médio por m²'] +
    media_por_cidade['Condomínio Médio por m²']
)
display(media_por_cidade)
```

```python
media_pet = media_por_cidade.loc[(slice(None), 'aceita'), 'Total Médio por m²'].mean()
print(f'Total Médio por m² para imóveis que aceitam pets é {media_pet:.2f}')

media_nao_pet = media_por_cidade.loc[(slice(None), 'nao aceita'), 'Total Médio por m²'].mean()
print(f'Total Médio por m² para imóveis que não aceitam pets é {media_nao_pet:.2f}')
```

### 3.	Analisar relação entre as variáveis explicativas quantitativas com preço do aluguel 

```python
df.groupby('comodos').aluguel.describe()

#Como o número de cômodos afeta o aluguel?
#    quanto mais comodos um imovel tem, mais caro seu aluguel tente a ser
```

```python
mc=df.groupby('comodos')['aluguel'].mean()
mec=mc.tolist()

comodos=np.arange(1,6,1)

plt.bar(comodos, mec, color='skyblue')
plt.xlabel('vagas')
plt.ylabel('Preço Médio do Aluguel')
plt.title('Influência dos comodos no Preço do Aluguel')
plt.xticks(comodos)
plt.grid(axis='y')

# Mostrar o gráfico
plt.show()
```

```python
df.groupby('banheiros').aluguel.describe()

#O número de banheiros influencia no preço do aluguel?
    #sim, tende a ser mais caro, a queda do grafico provavelmente se deve a menor amosta de informações
```

```python
mb=df.groupby('banheiros')['aluguel'].mean()
meb=mb.tolist()

banheiros=np.arange(1,8,1)

plt.bar(banheiros, meb, color='skyblue')
plt.xlabel('banheiros')
plt.ylabel('Preço Médio do Aluguel')
plt.title('Influência dos banheiros no Preço do Aluguel')
plt.xticks(banheiros)
plt.grid(axis='y')

# Mostrar o gráfico
plt.show()
```

```python
df.groupby('vagas').aluguel.describe()

#A presença de vagas de garagem está associada a aluguéis mais altos?
# quanto mais vagas mais caro tende a ser o aluguel do imovel
```

```python
mv=df.groupby('vagas')['aluguel'].mean()
mev=mv.tolist()

vagas=np.arange(0,7,1)

plt.bar(vagas, mev, color='skyblue')
plt.xlabel('vagas')
plt.ylabel('Preço Médio do Aluguel')
plt.title('Influência das vagas no Preço do Aluguel')
plt.xticks(vagas)
plt.grid(axis='y')

# Mostrar o gráfico
plt.show()
```

```python
df.groupby('andar').aluguel.describe()
#Andares mais altos são mais caros?
    #sim, a media de preço tende a crescer, e a queda no grafico provavelmente se deve a menor amostra de dados de predios com 29 andares
```

```python
mean_values = df.groupby('andar')['aluguel'].mean()
mean = mean_values.tolist()


andar=np.arange(0,30, 1)

plt.figure(figsize=(18, 9))
plt.bar(andar, mean, color='skyblue')
plt.xlabel('andar')
plt.ylabel('Preço Médio do Aluguel')
plt.title('Influência do andar no Preço do Aluguel')
plt.xticks(andar)
plt.grid(axis='y')
```

### 4.	Analisar relação entre área e valor do aluguel para imóveis mobiliados e não mobiliados 

```python
df_mobiliado = df[df['mobiliado']=='mobiliado'] #df com imóveis mobiliados
df_nao_mobiliado = df[df['mobiliado']=='nao mobiliado'] #df com imóveis não mobiliados
```

[Volta ao Índice](#indice)

```python
#Correlação entre aluguel e área para cada tipo de imóvel, pergunta 1
crr_mob = df_mobiliado.area.corr(df_mobiliado.aluguel)
crr_n_mob = df_nao_mobiliado.area.corr(df_nao_mobiliado.aluguel)
print(f'Correlação entre aluguel e área nos imóveis mobiliados: {crr_mob}\nCorrelação entre aluguel e área nos imóveis não mobiliados: {crr_n_mob}')
```

```python
#Gráfico de dispersão mobiliado
plt.scatter(df_mobiliado.area, df_mobiliado.aluguel, c='red', alpha=0.4)
plt.title('Área x Aluguel (imóveis mobiliados)')
plt.xlabel('Área')
plt.ylabel('Aluguel')
plt.show()

#Gráfico de dispersão não mobiliado
plt.scatter(df_nao_mobiliado.area, df_nao_mobiliado.aluguel, c='red', alpha=0.4)
plt.title('Área x Aluguel (imóveis não mobiliados)')
plt.xlabel('Área')
plt.ylabel('Aluguel')
plt.show()
```
